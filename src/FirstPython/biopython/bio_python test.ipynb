```python
from Bio.Seq import Seq

conding_dna=Seq("ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG")
template_dna=conding_dna.complement()
print(f'coding DNA :{conding_dna}')
print(f'template DNA : {template_dna}')

transribe_to_rna=conding_dna.transcribe()
print(f'trascribe : {transribe_to_rna}')
translate_to_protein=transribe_to_rna.translate(to_stop=True)
print(f'translate : {translate_to_protein}')
```

```python
from Bio.Seq import Seq

def mask_uncertain_bases(seq, quality_scores, threshold):
    return Seq(''.join('N' if score < threshold else base 
                       for base, score in zip(seq, quality_scores)))

seq = Seq("ATCGATCGATCG")
quality_scores = [30, 40, 20, 10, 25, 30, 35, 5, 15, 40, 40, 40]
masked_seq = mask_uncertain_bases(seq, quality_scores, 20)
print(masked_seq)  # 출력: ATNNNTNNAATCG
```
